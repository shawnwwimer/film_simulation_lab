# Basic Python
We're using Jupyter Notebook for this lab. Jupyter makes it easy to visualize and run the code in separate parts. A notebook is separated into individual cells with the input code and the output from running a cell is put immediately after that cell.

## Variables
Variables can represent virtually anything in Python; we'll use them mostly for values. Variable names must start with a letter and can contain numerals and underscores ( \_ ). Generally their names should be a description of what they are for. Here are some examples of correct and incorrect variable names

<code>L = 256     # fine
L1 = 256    # fine
L_ = 256    # fine
1L = 256    # wrong
</code>

## Comments
Everything following '#' in a line is a comment. They have no effect on the operation of the code and are used to describe how it works and why things work. You should use them liberally. Comments can also be made by enclosing text in triple quotes, like:

<code> ''' This is another comment '''
''' 
    Comments like this can be separated 
    over multiple lines 
'''
</code>

## Number types
We will deal with two types of numbers: integers and floats. Integers represent integers while floats represent numbers that can have decimal components. It is unlikely to matter in most cases and in Python you don't need to specify type.

## Mathematical operators
Basic operators are as follows
* $x+y$: <code>x + y</code>
* $x-y$: <code>x - y</code>
* $\frac{x}{y}$: <code>x / y</code>
* $x^y$: <code>x**y</code>
* $x \mod y$: <code>x % y</code>

Interestingly, the result of the modulo operation is restricted to the sign of $y$ and shouldn't be used for floats. Instead, you'd use a function.

## Functions
Functions are named similarly to variables and represent processes. Sometimes they take arguments, sometimes they don't. Sometimes they 'return' value(s) for further use, sometimes they don't. Sometimes they have side-effects, sometimes they don't. A simple function used often is <code>print()</code>, which is usually used with one argument. Using some variables, operators, and functions:

<code>x = 2
y = x**2
z = y * 2 / 3
print(z)
</code>
$z$ will now have the value 2.6666666666666665 and <code>print(z)</code> will print it in the cell's output.

## Strings
Strings are ordered collections of characters and can also be stored in a variable. They are declared by enclosing the collection with single quotes or double quotes (but not both!) like so:

<code>s = 'string'   # fine
s = "string"   # fine
s = 'string"   # wrong
</code>
If you want to include a backslash (\), you need to 'escape' it by using another backslash:
<code>s = '\\'</code>
When we use Matplotlib we have access to certain LaTeX commands, like '\theta' or '\cos' that will also need to be escaped.

A useful kind of string, particularly for printing out results, is the f-string. The f-string makes for easy formatting of a string with variables or operations. For example,
<code>x = 2
print(f'The square of {x} is {x**2}.')
</code>
This will print the line:
The square of 2 is 4.
Now we can change the value of $x$ and the print statement will be changed accordingly. Note that the 'f' comes before the string and the values are enclosed in curly brackets.



Try running the cell below to see how Python and Jupyter work.

```python
''' Example code from above '''
x = 2 
y = x**2
z = y * 2 / 3
print(z)

print(f'The square of {x} is {x**2}.')
```

# More basic Python
## Lists
Lists are collections of objects, which can include strings and numbers. Lists are declared simply by enclosing comma-separated objects within square brackets, like so:

<code>L = [0, 1, 2]</code>

An object is appended to a list like so: <code>L.append(3)</code>. Items in the list are zero-indexed, so <code>L[0]</code> is the first item and <code>L[n]</code> is the $n-1$ item. A neat feature in Python is that the items can be accessed from the end with negative numbers, so if <code>L</code> is $N$ items long, then <code>L[-1]</code> will get the $N-1$ item (which is also the last) and <code>L[x]</code> will get the $N-x$ item, which is $x-1$ from the end. You can't get items outside the bounds of the list; an error will occur and the execution will stop. The length of the list can be gotten from a function: <code>len(L)</code>.

Lists can be sliced, which returns a new list of selected elements from the original. For instance <code>L[1:4]</code> will return a list with the second through fourth elements.

## Tuples
Tuples are similar to lists in that they are collections of objects. Tuples are declared similarly by enclosing comma-separated objects within parentheses, like so:
<code>T = (0, 1, 2)</code>
The primary difference between lists and tuples is that tuples cannot be changed after they are created (they are 'immutable').

## Control statements
Control statements are used to control the order of code execution. There are three common control-flow statements,
* <code>if</code> \[condition\]<code>:</code>
* <code>while</code> \[condition\]<code>:</code>
* <code>for</code> \[item\] <code>in</code> \[collection\]<code>:</code>

An if-statment checks if a condition is true or false; only if it is true will the code inside execute. A while-statement checks the condition, then runs the code if it's true, and repeats this process until the condition is false. A for-statement executes once for each item in the collection and helpfully exposes the item.

White-space at the beginning of the line is important because Python uses it to group lines in control statements. You need it to be consistent; most code editors will do it automatically but you'll want to be careful.

Check some examples below.

```python
''' Example for lists and control statments '''
even = [] # create an empty list
odd = [] # create an odd list
for i in range(10): # range(x) will make a collection of integers from 0 to x-1
    if i % 2: # check if i is odd
        odd.append(i)
    else: # otherwise i is odd
        even.append(i)

for o in odd: # iterate over odd
    print(o) # print the item in odd
print(odd) # you can print the entire list together
print(odd[1:4]) # print the second through fourth elements

length = len(even) # get length of list
for i in range(length): # iterate from 0 to length
    print(f'Item {i} in even is {even[i]}') # print a formatted list
```

# Basic NumPy
Numpy is a commonly used module for making computations easier, particular with matrices or other collections of numbers. It offers many built-in mathematical operations and a useful interface for random numbers. Like other modules, the module is included in code with an <code>import</code> statement:

<code>import numpy as np</code>

## Arrays
We'll need to use arrays to represent film heights or other quantities. While arrays can be specified explicitly, some useful array-creation functions are:

* <code>np.zeros(shape)</code>: array of dimensions $shape$ filled with 0
* <code>np.ones(shape)</code>: array of dimensions $shape$ filled with 1
* <code>np.linspace(start, stop, number)</code>: $num$ evenly-spaced numbers between start and stop
* <code>np.arange(start, stop, step)</code>: samples spaced $step$ from each other between start and stop

The $shape$ is the dimensions of the array, so if <code>shape = (1,2,3)</code> then the array is $1\times2\times3$. The values in $shape$ have to be integers, floats will throw errors. There are many things one can do with the shape of an array but they won't be necessary in this lab.

You access elements in arrays like lists, similarly zero-indexed. <code>A[1,2]</code> returns the item in the second row and third column. You can access 'views' of the list by slicing: <code>A[:,2]</code> gets all rows in the third column.

A mathematical function is applied elementwise to the entire array and returns a new array, so either the sizes need to match or one needs to be a single value.

## Random numbers
NumPy has comprehensive random number generators but the process can be a little confusing at first. <code>numpy.random</code> is the submodule that contains what we'll need. The first step is to create a random number generator. A generator can be given a specific seed if a consistent sequence is desired, which will be useful for comparing similar methods.

After the generator is created it can be sampled according to one of the many distribution functions NumPy offers. The sampling can occur one at a time or for entire arrays if a shape is passed to the function. Check the examples below to see the process.

```python
''' Random number example '''

import numpy as np # must import before use

rng = np.random.default_rng() # create the random number generator

x = rng.uniform() # get a single number in [0,1) according to a uniform distribution
print(x)
A = rng.choice([-1.5, 0.25, 0.75], size=(3,3)) # create a 3x3 array randomly chosen from the supplied list
print(A) # show the entire array
print(A[1:,1:]) # show a slice of the array
A[1:,1:] = rng.integers(0, 5, size=(2,2)) # replace the slice with integers in [0,5), uniformly distribution
print(A)
```

If we create $A$ with <code>integers()</code> and then replace the slice with <code>choice()</code> we get some unexpected behavior.

```python
''' What if we swapped the choice and integers calls? '''
# don't need to import numpy again if it's already been imported in an executed cell

rng = np.random.default_rng() # create the random number generator

A = rng.integers(0, 5, size=(3,3)) # create a 3x3 array with integers in [0,5), uniformly distribution
print(A) # show the entire array
print(A[1:,1:]) # show a slice of the array
A[1:,1:] = rng.choice([-1.5, 0.25, 0.75], size=(2,2)) # replace the slice with numbers chosen from the supplied list
print(A)
```

Because $A$ was created with <code>integers()</code> then A is initialized as an integer array. Notice the decimal points after numbers in the previous example compared to here. However, <code>choice()</code> creates an array of floats before trying to copy them into $A$. Because $A$ is an integer array, the floats are truncated into integers before reassigning $A$. Normally this will not be an issue but it is something to watch for if you can't tell what's going wrong.

# Matplotlib basics
Matplotlib is a widely used module for graphing, plotting, and figure creation. It was initially designed to be similar to MATLAB usage. We'll mostly use multi-plot 2D figures which has a fairly standard operation. I find it easier to look at examples for Matplotlib, let's use it to check the random number distributions from NumPy.

```python
''' Are the distribtuions good? '''

from matplotlib import pyplot as plt # matplotlib has many submodules, we'll use pyplot

rng = np.random.default_rng() # create the random number generator

samples = 1000000
nbins = 50

uniform = rng.uniform(size=samples)
normal = rng.normal(size=samples)

# create some expected distribution shapes
x1 = np.linspace(0, 1, 100) # range is [0,1)
uniform_line = np.ones(x1.size) # PDF is 1 in [0,1)
x2 = np.linspace(-5, 5, 100)
normalization = 1/np.sqrt(2*np.pi) # numpy offers a square root function and an easy pi
normal_curve = normalization * np.exp(-x2**2/2) # numpy also offers an exponential function

fig, ax = plt.subplots(1,2, figsize=(6,3), dpi=100) # create a figure with 1 row and 2 columns of axes; figsize and dpi to look nicer

# note that ax is a collection of axes
ax[0].hist(uniform, bins=nbins, density=True) # histogram in the first axis of uniform distribution
ax[1].hist(normal, bins=nbins, density=True) # histogram in the second axis of normal/Gaussian distribution

# we can plot curves onto the same axes
ax[0].plot(x1, uniform_line, linewidth=2) # x1 and uniform_line must be the same shape
ax[1].plot(x2, normal_curve, linewidth=2)

# create axis labels
ax[0].set_ylabel('Uniform distribution', fontsize=12)
ax[1].set_ylabel('Normal distribution', fontsize=12)
ax[0].set_xlabel('x', fontsize=12)
ax[1].set_xlabel('x', fontsize=12)

# create legends from lists of strings in the right order
ax[0].legend(['Counts', 'PDF'])
ax[1].legend(['Counts', 'PDF'], loc='lower right')

fig.tight_layout() # tries to fit the plots together nicely
plt.show() # displays the figure
```

```python
''' Plot a 2D function '''

rng = np.random.default_rng() # create the random number generator

shape = (100, 100, 100) 

uniform = rng.uniform(size=shape) # create a 100x100x100 array of floats in [0,1)
sum_of_uniform = np.sum(uniform, axis=2) # sum the entire array along the z-direction, collapsing it into 2 dimensions

fig, ax = plt.subplots(figsize=(6,3), dpi=100) # only need one axis
image = ax.imshow(sum_of_uniform, origin='lower') # imshow creates a 2D image
plt.colorbar(image, ax=ax) # add a colorbar according to image on ax


# create axis labels
ax.set_ylabel('y', fontsize=12)
ax.set_xlabel('x', fontsize=12)

fig.tight_layout() # tries to fit the plots together nicely
plt.show() # displays the figure
```
